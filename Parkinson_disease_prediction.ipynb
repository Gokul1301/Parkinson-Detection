## Scaling the Input

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = df.drop(['status'],axis=1)
y = df['status']
```

```python
std_scaler = StandardScaler()


x = std_scaler.fit_transform(x)
```

```python
x
```

# train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
print(y_train.shape)
```

```python
print(y_test.shape)
```

```python
x_train = np.array(x_train)
```

```python
y_train = np.array(y_train)
```

```python
x_test = np.array(x_test)
```

```python
y_test = np.array(y_test)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
```

```python
model1.fit(x_train,y_train)
```

```python
y_premodel1 = model1.predict(x_test)
y_premodel1
```

```python
for i,j in enumerate(y_premodel1):
  if(j<0.5):
    y_premodel1[i]=0
  else:
      y_premodel1[i]=1
      print(y_premodel1)
      y_premodel1.shape

```

```python
y_test = np.array(y_test)
print(y_test)
y_test.shape
```

```python
print('x_test_array   y_test')
for i, j in zip(y_premodel1, y_test):
    print(i,j)
```

```python
print('linear regression accuracy :\n')
model1.score(x_test,y_test)
```

# Neural Network

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model7 = Sequential()
model7.add(Dense(1, input_dim=22, activation='sigmoid'))
model7.add(Dense(20, activation='sigmoid'))
model7.add(Dense(12, activation='sigmoid'))
model7.add(Dense(9, activation='sigmoid'))
model7.add(Dense(5, activation='sigmoid'))
model7.add(Dense(1, activation='sigmoid'))

model7.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model7.fit(x_train, y_train, epochs=1000)
```

```python
model7.evaluate(x_test, y_test)
```

```python
pred_nn = model7.predict(x_test)

for i, j in enumerate(pred_nn):
    if j < 0.5:
        pred_nn[i] = 0
    else:
        pred_nn[i] = 1

pred_nn = pred_nn.flatten()
print(pred_nn)
print(y_test)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming y_test and y_predmod1 are already defined
cn = confusion_matrix(y_test, pred_nn)  # Assuming pred_nn is the predicted values from your model
plt.figure(figsize=(8, 6))
fg = sns.heatmap(cn, annot=True, fmt='d')  # Corrected the formatting to display integers
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()  # Added plt.show() to display the plot
```
